```python
import numpy as np # Mengimpor Library NumPy untuk menyelesaikan operasi matematika

def integrate_rec(f, a, b, n): # Alat/parameter untuk menghitung luas dimana f (fungsi), a (batas bawah/batas awal), b (batas atas/batas akhir), n (jumlah sub-interval/jumlah kotak)
    # Implementation of the rectanlge method 
    h = (b-a)/n # menghitung lebar setiap sub-interval (kotak)
    x = np.linspace(a, b, n+1) # Membuat daftar titik-titik pembagi pada sumbu x dari a sampai b
    i=0 # Menentukan nilai awal urutan pengerjaan mulai dari 0
    area=0 # Menentukan nilai awal total luas mulai dari 0
    while i<n: # Melakukan pengulangan hitungan selama urutan belum mencapai jumlah kotak
        sup_rect =  f(x[i])*h # Menghitung luas elemen/kotak: tinggi f(x) pada sisi kiri dikalikan lebar h
        area += sup_rect # Menambahkan luas elemen/kotak ke dalam total luas keseluruhan
        i += 1 # Menambah urutan pengerjaan untuk lanjut ke elemen/kotak berikutnya
    return area # Memberikan hasil akhir total luas yang sudah dikumpulkan
 # Memberikan tanda 'r' agar karakter khusus (\) dibaca sebagai teks biasa
r'''
We test the Rectangle method on the sine funcion were the definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''
# Menjalankan fungsi dengan variasi jumlah kotak (5, 10, dan 100)
sup_5 = integrate_rec(np.sin, 0, np.pi/2, 5)
sup_10 = integrate_rec(np.sin, 0, np.pi/2, 10)
sup_100 = integrate_rec(np.sin, 0, np.pi/2, 100)

# Menampilkan hasil dengan batasan dua angka di belakang koma
print('Using n=5, the rectangle method returns a value of {:.2f}'.format(sup_5))
print('Using n=10, the rectangle method returns a value of {:.2f}'.format(sup_10))
print('Using n=100, the rectangle method returns a value of {:.2f}'.format(sup_100))
     
''' 
Output:
Using n=5, the rectangle method returns a value of 0.83
Using n=10, the rectangle method returns a value of 0.92
Using n=100, the rectangle method returns a value of 0.99
'''
 
```

```python
import numpy as np # Mengimpor Library NumPy untuk menyelesaikan operasi matematika

def integrate_trap(f, a, b, n): # Alat/parameter untuk menghitung luas dimana f (fungsi), a (batas bawah/batas awal), b (batas atas/batas akhir), n (jumlah sub-interval/jumlah kotak)
    # Implementation of the trapezoidal rule
    h = (b-a)/n # Menghitung lebar setiap bagian
    x = np.linspace(a, b, n+1) # Membuat daftar titik-titik pembagi pada sumbu x dari a sampai b
    i=1 # Menentukan urutan pengerjaan dimulai dari kotak ke-1
    area = h*(f(x[0]) + f(x[n]))/2 # Menghitung nilai awal untuk area menggunakan rumus ujung awal dan ujung akhir
    # (f(awal) + f(akhir)) dikali lebar dibagi 2
    while i<n: # Melakukan pengulangan hitungan selama urutan belum mencapai jumlah kotak
        sup_rect = f(x[i])*h # Menghitung kontribusi nilai fungsi di titik tengah dikali lebar
        area += sup_rect # Menambahkan hasil tersebut ke dalam total luas keseluruhan
        i += 1 # Menambah urutan untuk lanjut ke titik berikutnya
    return area # Memberikan hasil akhir total luas yang sudah dikumpulkan

# Tanda 'r' digunakan agar karakter khusus (\) dibaca sebagai teks biasa
r'''
We test the trapezoidal rule on the known sine funcion were the  
definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''
# Menjalankan fungsi dengan variasi jumlah bagian (5 dan 10)
sup_5 = integrate_trap(np.sin, 0, np.pi/2, 5)
sup_10 = integrate_trap(np.sin, 0, np.pi/2, 10)

# Menampilkan hasil di layar dengan batasan dua angka di belakang koma
print('Using n=5, the trapezoidal rule returns a value of {:.2f}'.format(sup_5))
print('Using n=10, the trapezoidal rule returns a value of {:.2f}'.format(sup_10))
   
'''
Output:
Using n=5, the trapezoidal rule returns a value of 0.99
Using n=10, the trapezoidal rule returns a value of 1.00
'''
 
```

```python
import numpy as np # Mengimpor Library NumPy untuk menyelesaikan operasi matematika
from scipy import integrate # Mengambil alat hitung integral otomatis dari library SciPy


x = np.linspace(0,9, 3) # 3 divisions %*\textcolor{codegreen}{$[x_0,x_1,x_2]$}*), n=2 # Membuat daftar titik pada sumbu x dari 0 sampai 9, dibagi menjadi 3 titik (2 bagian)
y = x**2 # Menentukan fungsi matematika y = x pangkat 2

sup_trapz = integrate.trapezoid(y,x) # Menggunakan alat otomatis untuk menghitung dengan metode Trapesium
sup_simps = integrate.simpson(y,x) # Menggunakan alat otomatis untuk menghitung dengan metode Simpson
# Metode Simpson biasanya jauh lebih akurat untuk garis yang sangat melengkung

# Menampilkan hasil hitungan tanpa angka di belakang koma
print('Using n=2, the trapezoidal rule returns a value of {:.0f}'.format(sup_trapz))
print('Using n=2, the composite Simpson rule returns a value of {:.0f}'.format(sup_simps))

'''
Output:
Using n=2, the trapezoidal rule returns a value of 273
Using n=2, the composite Simpson rule returns a value of 243
'''
```

```python
import numpy as np # Mengimpor Library NumPy untuk mengolah deretan angka 
from scipy.integrate import trapezoid # Mengimpor alat untuk menghitung luas dari Library SciPy
import matplotlib.pyplot as plt # Mengimpor Library Matplotlib untuk membuat grafik visual 

r = np.linspace(1,6400,6400) # Membuat daftar jarak (r) dari pusat Bumi ke permukaan (1 km sampai 6400 km)

def density(): # Mendefinisikan nilai kerapatan massa pada setiap lapisan Bumi:
               # Inti dalam, inti luar, mantel bawah, mantel atas, dan kerak.
    ro_inner_core = np.linspace(13100, 12800, 1220) # Nilai inti dalam bumi
    ro_outer_core = np.linspace(12200, 9900, 2259) # Nilai inti luar bumi
    ro_lower_mantle = np.linspace(5600,4400,2171) # Nilai mantel bawah bumi
    ro_upper_mantle = np.linspace(4130,3400,720) # Nilai mantel  atas bumi
    ro_crust = np.linspace(3100,2700,30) # Nilai dari kerak bumi
   
    ro_final = np.concatenate((ro_inner_core, ro_outer_core, ro_lower_mantle, ro_upper_mantle, ro_crust)) # Menggabungkan semua data lapisan bumi menjadi satu kesatuan yang utuh
    
    return ro_final # Memberikan hasil akhir berupa daftar kerapatan massa dari pusat ke permukaan bumi
  
ro = density() # Menjalankan fungsi dan menyimpan hasilnya dalam variabel 'ro'

# Membuat Grafik
fig, ax = plt.subplots() # Membuat bingkai grafik
ax.plot(r, ro, label="Densities on the Earth's Interior") # Menggambar garis grafik
ax.set_ylabel(r"Density [Kg/m$^3$]")  # Memberi label pada sumbu tegak (Kerapatan)
ax.set_xlabel("Distance on the Earth center r [km]") # Memberi label pada sumbu mendatar (Jarak)
ax.legend() # Menampilkan kotak keterangan label
ax.grid() # Menampilkan garis-garis kotak (grid) bantuan pada latar belakang
```

```python
def gravity(r): # Mendefiniskan parameter
    
    g = np.zeros(len(r)) # Menyiapkan daftar angka nol sebagai tempat menyimpan hasil gravitasi
    Gr = 6.67408e-11 # Menentukan konstanta gravitasi umum (G)
    r = r * 1000 # from Km to m # Mengubah satuan jarak r dari Kilometer (km) menjadi Meter (m)
    
    for i in range(1,len(r)): # Melakukan pengulangan hitungan untuk setiap titik jarak r
       
        r1 = r[0:i]  # Mengambil potongan data jarak dan densitas dari pusat sampai titik i 
        ro1  = ro[0:i]  # Mengambil potongan data jarak dan densitas dari pusat sampai titik i 
        r2 = r1[i-1] # Menentukan jarak dari pusat pada titik yang sedang dihitung
    
        y = ro1*r1**2 # Menghitung nilai fungsi di dalam integral (densitas dikali kuadrat jarak)
        y_int = trapezoid(y,r1) # Melakukan integrasi otomatis menggunakan metode Trapesium
        
        g1 = ((4 * np.pi*Gr)/(r2**2)) * y_int # Menghitung nilai gravitasi berdasarkan rumus hukum gravitasi Newton
        g[i] = g1 # Menyimpan hasil hitungan ke dalam daftar g pada urutan ke-i
           
    return g # Memberikan hasil akhir berupa daftar nilai gravitasi

g = gravity(r) # Menjalankan fungsi untuk mendapatkan nilai gravitasi

# Membuat Grafik
fig, ax = plt.subplots() # Membuat bingkai grafik
ax.plot(r,g) # Menggambar grafik hubungan jarak (r) dengan gravitasi (g)
ax.grid() # Menampilkan garis-garis kotak (grid) bantuan pada latar belakang
ax.set_ylabel(r'g [m/s$^2]$') # Memberi label sumbu tegak (Percepatan Gravitasi)
ax.set_xlabel('Distance from the Earth center r [km]') # Memberi label pada sumbu mendatar (Jarak)
```

```python
def pressure(r, ro, g): # Mendefinisikan parameter berupa nilai jarak, kerapatan massa, dan gravitasi bumi
    
    p = np.zeros(len(r)) # Menyiapkan wadah kosong (angka nol) untuk menyimpan hasil tekanan
    r = r *1000 # Mengubah satuan jarak r dari kilometer ke meter
    
    for i in range(0,len(r)): # Melakukan pengulangan hitungan untuk setiap titik kedalaman  
     # Mengambil potongan data jarak, densitas, dan gravitasi dari titik yang digitung ke permukaan
        r1 = r[i:len(r)]
        ro1 = ro[i:len(r)]
        g1 = g[i:len(r)]

        y = ro1*g1 # Menghitung fungsi y (densitas dikali gravitasi)
        p1 = trapezoid(y,r1) # Menghitung akumulasi tekanan menggunakan metode Trapesium
        p[i] = p1 # Menyimpan hasil tekanan ke dalam daftar p
    return p # Mengembalikan daftar nilai tekanan
    
p = pressure(r,ro,g)/1e9 # expressed in GPa # Menjalankan fungsi dan membagi dengan 1e9 (satu miliar) agar satuannya menjadi GigaPascal (GPa)
z = np.linspace(6400, 1, 6400) # Membuat daftar kedalaman z (dari permukaan 6400 km menuju pusat 1 km)

# Membuat Grafik
fig, ax = plt.subplots() # Membuat bingkai grafik
ax.plot(z,p) # Menggambar grafik hubungan kedalaman (z) dengan tekanan (p)
ax.grid() # Menampilkan garis-garis kotak (grid) bantuan pada latar belakang
ax.set_ylabel('P [GPa]') # Memberi label pada sumbu tegak (Tekanan dalam GigaPascal)
ax.set_xlabel('Depth z from the Earth Surface [km]') # Memberi label pada sumbu mendatar (Kedalaman dari permukaan)
```
