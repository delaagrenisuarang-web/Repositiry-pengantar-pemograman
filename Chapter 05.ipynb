```python
import pandas as pd # Mengimpor library Pandas untuk mengolah tabel data
from scipy.stats.mstats import gmean, hmean # Mengambil rumus rata-rata ukur (gmean) dan harmonis (hmean)
import matplotlib.pyplot as plt # Mengimpor library Matplotlib untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # Membaca file Excel 'Smith_glass_post_NYT_data.xlsx.' pada halaman 'Supp_traces' dan menyimpannya ke variabel 'my_dataset'

a_mean = my_dataset.Zr.mean() # Menghitung rata-rata hitung (aritmatika) dari kolom 'Zr' menggunakan fungsi bawaan Pandas
g_mean = gmean(my_dataset['Zr']) # Menghitung rata-rata ukur (geometris) dari kolom 'Zr' menggunakan library SciPy
h_mean = hmean(my_dataset['Zr']) # Menghitung rata-rata harmonis dari kolom 'Zr' menggunakan library SciPy

# Mencetak Hasil 
print ('-------')
print ('arithmetic mean')
print ("{0:.1f} [ppm]".format(a_mean)) # Mencetak hasil a_mean dengan 1 angka di belakang koma
print ('-------')

print ('geometric mean')
print ("{0:.1f} [ppm]".format(g_mean)) # Mencetak hasil g_mean dengan 1 angka di belakang koma
print ('-------')

print ('harmonic mean')
print ("{0:.1f} [ppm]".format(h_mean)) # Mencetak hasil h_mean dengan 1 angka di belakang koma
print ('-------')

# Memvisualisasikan Grafik
fig, ax = plt.subplots() # Membuat bingkai foto/grafik (fig) dan area gambar (ax)
ax.hist(my_dataset.Zr, bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=0.8)  # Membuat grafik Histogram dari data Zr
                                                                                                        # bins='auto': komputer menentukan sendiri jumlah batang yang pas
                                                                                                        # density=True: mengubah sumbu Y menjadi nilai peluang (0 sampai 1)
                                                                                                        # alpha=0.8: mengatur tingkat transparansi warna batang
# Menambahkan garis vertikal (axvline) untuk menandai posisi rata-rata pada grafik
ax.axvline(a_mean, color='red', label='Arithmetic mean', linewidth=2) # Garis mearh untuk aritmatika
ax.axvline(g_mean, color='pink',  label='Geometric mean', linewidth=2) # Garis merah muda untuk geometris
ax.axvline(h_mean, color='brown',  label='Harmonic mean', linewidth=2) # Garis cokelat untuk harmonis
ax.set_xlabel('Zr [ppm]') # Memberi nama label pada sumbu mendatar (X)
ax.set_ylabel('Probability density') # Memberi nama label pada sumbu tegak (Y)
ax.legend() # Menampilkan kotak keterangan warna garis dan batang (Legenda)

'''
Output:
-------
arithmetic mean
365.4 [ppm]
-------
geometric mean
348.6 [ppm]
-------
harmonic mean
333.8 [ppm]
-------
'''
```

```python
import pandas as pd  # Mengimpor library Pandas untuk membaca dan mengolah data tabel
import numpy as np # Mengimpor library NumPy untuk melakukan perhitungan matematika
import matplotlib.pyplot as plt # Mengimpor library Matplotlib untuk membuat visualisasi grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # Membaca data dari file Excel pada sheet 'Supp_traces'

hist, bin_edges = np.histogram(my_dataset['Zr'], bins= 20, density=True) # Membuat histogram data 'Zr' dengan 20 kategori (bins) secara matematis
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2 # Mencari nilai tengah dari bin (kelompok) yang memiliki frekuensi tertinggi (modus)

print ('modal value: {0:.0f} [ppm]'.format(modal_value)) # Mencetak hasil perhitungan modus ke layar dengan pembulatan 0 desimal

fig, ax = plt.subplots() # Membuat objek gambar (figure) dan sumbu (axis) untuk plot
ax.hist(my_dataset.Zr, bins=20, density=True, edgecolor='k', label="Measurements Hist", alpha=0.8) # Membuat grafik histogram visual dari kolom 'Zr'
ax.axvline(modal_value, color="orange", label="Modal value", linewidth=2) # Menambahkan garis vertikal berwarna oranye tepat di titik nilai modus
ax.set_xlabel('Zr [ppm]') # Memberikan label nama pada sumbu horizontal (X) 
ax.set_ylabel('Probability density') # Memberikan label nama pada sumbu vertikal (Y)
ax.legend() # Memberikan label nama pada sumbu horizontal (X) dan vertikal (Y)

'''
Output: modal value: 277 [ppm]
'''
```

```python
import pandas as pd # Mengimpor library Pandas untuk membaca dan mengolah data tabel
import matplotlib.pyplot as plt # Mengimpor library Matplotlib untuk membuat visualisasi grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # Membaca data dari file Excel pada sheet 'Supp_traces'

# Menghitung Varian (Variance) dan Standar Deviasi (Standard Deviation) data 'Zr'
variance =  my_dataset['Zr'].var()
stddev =  my_dataset['Zr'].std()

# Mencetak hasil Varian ke terminal
print ('-------')
print ('Variance')
print("{0:.0f} [square ppm]".format(variance)) # Hasil varians dalam satuan kuadrat (square)
print ('-------')

# Mencetak hasil Standar Deviasi ke terminal
print ('Standard Deviation')
print("{0:.0f} [ppm]".format(stddev)) # Hasil standar deviasi dalam satuan yang sama dengan data (ppm)
print ('-------')

fig, ax = plt.subplots() # Membuat objek gambar (figure) dan sumbu (axis) untuk plot
ax.hist(my_dataset.Zr, bins= 20, density = True, edgecolor='k', label='Measurements Hist') # Membuat grafik histogram dari data 'Zr' dengan 20 kategori (bins)
ax.axvline(my_dataset['Zr'].mean()-stddev, color='purple', label=r'mean - 1$\sigma$', linewidth=2) # Menambahkan garis vertikal ungu untuk batas bawah standar deviasi (Mean - 1 Sigma)
ax.axvline(my_dataset['Zr'].mean()+stddev, color='green', label=r'mean + 1$\sigma$', linewidth=2) # Menambahkan garis vertikal hijau untuk batas atas standar deviasi (Mean + 1 Sigma)
ax.axvspan(my_dataset['Zr'].mean()-stddev, my_dataset['Zr'].mean()+stddev, alpha=0.1, color='orange', label=r'mean $\pm$ 1$\sigma$') # Memberi warna latar oranye transparan di antara batas -1 sigma dan +1 sigma
ax.set_xlabel('Zr [ppm]')  # Memberikan label nama pada sumbu X 
ax.set_ylabel('Probability density')   # Memberikan label nama pada sumbu Y
ax.legend() # Menampilkan legenda di dalam grafik

'''
Output:
-------
Variance
14021 [square ppm]
-------
Standard Deviation
118 [ppm]
-------
'''
```

```python
import pandas as pd # Mengimpor library Pandas untuk membaca dan mengolah data tabel
import numpy as np # Mengimpor library NumPy untuk melakukan perhitungan matematika
import matplotlib.pyplot as plt  # Mengimpor library Matplotlib untuk membuat visualisasi grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # Membaca data dari file Excel pada sheet 'Supp_traces'

# Menghitung Kuartil 1 (persentil 25) dan Kuartil 3 (persentil 75)
q1 = np.percentile(my_dataset.Zr, 25, method= 'midpoint') 
q3 = np.percentile(my_dataset.Zr, 75, method= 'midpoint') 
  
iqr = q3 - q1 # Interquaritle range (IQR)  # Menghitung IQR (selisih antara Kuartil 3 dan Kuartil 1)

#Mencetak hasil perhitungan ke terminal
print ('-------')
print ('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr))
print ('-------')

fig, ax = plt.subplots() # Membuat objek gambar (figure) dan sumbu (axis) untuk plot
ax.hist(my_dataset.Zr, bins='auto', density=True, edgecolor='k', label='Measurements Hist') # Membuat grafik histogram (bins='auto' berarti jumlah batang ditentukan otomatis oleh sistem)
ax.axvline(q1, color='purple', label='Q1', linewidth=2) # Menambahkan garis vertikal ungu untuk menandai Kuartil 1 (Q1)
ax.axvline(q3, color='green', label='Q3', linewidth=2) # Menambahkan garis vertikal hijau untuk menandai Kuartil 3 (Q3)
ax.axvspan(q1, q3, alpha=0.1, color='green', label='Interquaritle range (IQR)') # Memberi warna latar hijau transparan di antara Q1 dan Q3 (wilayah IQR)
ax.set_xlabel('Zr [ppm]') # Memberikan label nama pada sumbu X 
ax.set_ylabel('Probability density')  # Memberikan label nama pada sumbu Y
ax.legend() # Menampilkan legenda di dalam grafik

'''
Output:
-------
Interquaritle range (IQR): 164 [ppm]
-------
'''
```

```python
import numpy as np # Mengimpor library NumPy untuk melakukan perhitungan matematika
from scipy.stats import skew # Mengimpor fungsi khusus untuk menghitung kemiringan (skewness)

# Menghitung statistik dasar: Rata-rata (mean), Median, Modus (modal), dan Standar Deviasi
a_mean = my_dataset.Zr.mean()
a_mean = my_dataset.Zr.mean() 
median = my_dataset.Zr.median()
hist, bin_edges = np.histogram(my_dataset['Zr'], bins=20, density=True)
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2
standard_deviation = my_dataset['Zr'].std()

a1 = (a_mean - modal_value) / standard_deviation # Menghitung Koefisien Skewness Pertama Pearson (menggunakan Modus)
a2 = 3 * (a_mean - median) / standard_deviation # Menghitung Momen Skewness Kedua Pearson (menggunakan Median)
g1 = skew(my_dataset['Zr']) # Menghitung Koefisien Skewness Fisher-Pearson (menggunakan fungsi bawaan scipy)

# Mencetak hasil perhitungan kemiringan ke layar terminal
print ('-------')
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1))
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2))
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1))
print ('-------')

'''
Output:
-------
Pearson's first coefficient of skewness: 0.74
Pearson's 2nd moment of skewness: 0.66
Fisher-Pearson's coefficient of skewness: 1.26
-------
'''
```
